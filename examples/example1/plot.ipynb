```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = np.load("./output/results.npy")

print(data.shape)
```

```python
out_log = data[-1].reshape(6, 50)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(3, 2, 1)
ax.plot(out_log[1, :]*1000)
ax.set_title("Cl")

ax = fig.add_subplot(3, 2, 2)
ax.plot(out_log[0, :])
ax.set_title("pH")

ax = fig.add_subplot(3, 2, 3)
ax.plot(out_log[2, :]*1000)
ax.set_title("Mg")

ax = fig.add_subplot(3, 2, 4)
ax.plot(out_log[3, :]*1000)
ax.set_title("Ca")

ax = fig.add_subplot(3, 2, 5)
ax.plot(out_log[4, :]*1000/1.8)
ax.set_title("Calcite")

ax = fig.add_subplot(3, 2, 6)
ax.plot(out_log[5, :]*1000/1.8)
ax.set_title("Dolomite")
plt.show()
```
